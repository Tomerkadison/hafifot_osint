# Polars Exercises 02 — Selecting & Subsetting

Now that we know the data, the next everyday task is **choosing the columns we actually need** — the building block of every Foundry transform. In Polars this is done with `select()` and **expressions** (`pl.col`), not with `[[...]]` bracket lists like in pandas.


**Data file:** `data/orders.csv`

---

### 1. Import Polars as `pl`, import the selectors module as `cs`, and read `data/orders.csv` into `orders`.

```python
import polars as pl
import polars.selectors as cs

orders = pl.read_csv("data/orders.csv")
orders.head()
```

### 2. Select only the columns `order_id`, `product_name`, `quantity`.

```python
orders.select("order_id", "product_name", "quantity")
```

### 3. Do the same selection, but this time using **expressions** (`pl.col(...)`).

```python
orders.select(pl.col("order_id"), pl.col("product_name"), pl.col("quantity"))
```

### 4. Select **all columns except** `quantity` and `unit_price`.

```python
orders.select(pl.exclude("quantity", "unit_price"))
```

### 5. **Rename** `region` to `sales_region` and `channel` to `sales_channel`.

```python
orders.rename({"region": "sales_region", "channel": "sales_channel"})
```

### 6. **Drop** the `discount_code` column.

```python
orders.drop("discount_code")
```

### 7. Select all columns, but with `quantity` and `unit_price` **first** (reorder columns).

```python
orders.select("quantity", "unit_price", pl.exclude("quantity", "unit_price"))
```

### 8. Using **selectors**, select all the **string** columns.

```python
orders.select(cs.string())
```

### 9. Using **selectors**, select all the **numeric** columns.

```python
orders.select(cs.numeric())
```

### 10. Get the `quantity` column **as a Series**.

```python
orders.get_column("quantity")
```

### 11. Get a **single value**: the `unit_price` of the very first row.

```python
orders.select("unit_price").item(0, 0)
```

### 12. Select all columns whose name **starts with `order`** (use a selector).

```python
orders.select(cs.starts_with("order"))
```

### 13. Select `unit_price`, but rename it to `price` in the result (use `.alias`).

```python
orders.select(pl.col("unit_price").alias("price"))
```

### 14. Select the **first 3 columns by position** (hint: `orders.columns`).

```python
orders.select(orders.columns[:3])
```

### 15. Find all the **distinct combinations** of `region` and `channel`.

```python
orders.select("region", "channel").unique().sort("region", "channel")
```
