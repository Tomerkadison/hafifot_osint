# Polars Exercises 03 — Filtering & Sorting

Filtering rows is the Polars version of Excel's *quick filter* and pandas' boolean masks. We keep only the rows we care about with `filter()`, then put them in order with `sort()`. Master combining conditions with `&` (and), `|` (or), `~` (not) — it is the heart of every WHERE clause in our pipelines.

**Data file:** `data/orders.csv`

---

> **Solutions version** — try it yourself first from `Exercises.ipynb`.

### 1. Import Polars as `pl` and read `data/orders.csv` into `orders`.

```python
import polars as pl

orders = pl.read_csv("data/orders.csv")
orders.head()
```

### 2. Keep only the rows where `status` is `"completed"`.

```python
orders.filter(pl.col("status") == "completed")
```

### 3. Keep only orders with `quantity` greater than 20.

```python
orders.filter(pl.col("quantity") > 20)
```

### 4. Keep online orders **from Europe** (`region == "Europe"` **and** `channel == "online"`). Remember the parentheses around each condition.

```python
orders.filter((pl.col("region") == "Europe") & (pl.col("channel") == "online"))
```

### 5. Keep orders whose `category` is **either** `"pens"` or `"books"` (use `is_in`).

```python
orders.filter(pl.col("category").is_in(["pens", "books"]))
```

### 6. Keep orders where `quantity` is **between 10 and 20** (inclusive). Use `is_between`.

```python
orders.filter(pl.col("quantity").is_between(10, 20))
```

### 7. Find all the **returns** (rows where `quantity` is negative).

```python
orders.filter(pl.col("quantity") < 0)
```

### 8. Keep orders whose `product_name` **contains** the text `"Pen"` (use `str.contains`).

```python
orders.filter(pl.col("product_name").str.contains("Pen"))
```

### 9. Keep only rows that **have** a `discount_code` (not null).

```python
orders.filter(pl.col("discount_code").is_not_null())
```

### 10. Keep only rows that are **missing** a `discount_code` (null).

```python
orders.filter(pl.col("discount_code").is_null())
```

### 11. Keep every order **except** the cancelled ones (use `~` or `!=`).

```python
orders.filter(pl.col("status") != "cancelled")
```

### 12. Sort the whole table by `unit_price`, cheapest first.

```python
orders.sort("unit_price")
```

### 13. Sort by `quantity`, **largest first**.

```python
orders.sort("quantity", descending=True)
```

### 14. Sort by `region` (A→Z), and within each region by `unit_price` (high→low).

```python
orders.sort("region", "unit_price", descending=[False, True])
```

### 15. Show the **5 most expensive orders** (highest `unit_price`).

```python
orders.top_k(5, by="unit_price")
```

### 16. **How many** completed online orders from Europe have `quantity` of at least 15? (filter, then count the rows)

```python
orders.filter(
    (pl.col("status") == "completed")
    & (pl.col("channel") == "online")
    & (pl.col("region") == "Europe")
    & (pl.col("quantity") >= 15)
).height
```
