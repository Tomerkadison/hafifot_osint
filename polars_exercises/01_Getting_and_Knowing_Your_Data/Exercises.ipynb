# Polars Exercises 01 — Getting & Knowing Your Data

A raw `orders` dataset just landed in our pipeline (think of a new dataset in Palantir Foundry). Before we build *any* transform on top of it, we have to **get to know the data**: its shape, its columns, its types, and where it is dirty.

*הגיע אלינו dataset גולמי של הזמנות (כמו dataset חדש ב-Foundry). לפני שכותבים טרנספורמציה כלשהי — חייבים קודם להכיר את המידע: כמה שורות, אילו עמודות, אילו טיפוסים, ואיפה המידע “מלוכלך”.*

**Data file:** `data/orders.csv`

### 1. Import Polars under the alias `pl`.

*ייבאו את Polars תחת השם `pl`.*

### 2. Read `data/orders.csv` into a DataFrame called `orders`.

*קראו את הקובץ `data/orders.csv` ל-DataFrame בשם `orders`.*

### 3. Show the **first 10 rows**.

*הציגו את 10 השורות הראשונות.*

### 4. Show the **last 5 rows**.

*הציגו את 5 השורות האחרונות.*

### 5. How many **rows and columns** does the dataset have?

*כמה שורות ועמודות יש ב-dataset?*

### 6. What are the **column names**?

*מהם שמות העמודות?*

### 7. What is the **data type of every column**? Notice which type `order_date` got.

*מהו הטיפוס של כל עמודה? שימו לב איזה טיפוס קיבלה `order_date`.*

### 8. Get **summary statistics** for all columns in one call.

*קבלו סטטיסטיקה מסכמת לכל העמודות בקריאה אחת.*

### 9. Use **`glimpse()`** for a transposed, one-line-per-column overview (great for wide datasets).

*השתמשו ב-`glimpse()` לתצוגה הפוכה, שורה לכל עמודה.*

### 10. Select **only the `product_name` column**.

*בחרו רק את העמודה `product_name`.*

### 11. Select the columns `order_id`, `product_name`, `quantity`, `unit_price`.

*בחרו את העמודות `order_id`, `product_name`, `quantity`, `unit_price`.*

### 12. How many **distinct products** (`product_sku`) appear in the data?

*כמה מוצרים שונים (`product_sku`) מופיעים במידע?*

### 13. List the **distinct categories**.

*הציגו את הקטגוריות השונות.*

### 14. How many orders are there **per `status`**? (use `value_counts`)

*כמה הזמנות יש לכל `status`? (השתמשו ב-`value_counts`)*

### 15. Which **category** has the most orders?

*לאיזו קטגוריה יש הכי הרבה הזמנות?*

### 16. What are the **minimum and maximum `quantity`**? (remember: returns are negative)

*מהם הערך המינימלי והמקסימלי של `quantity`? (זכרו: החזרות הן שליליות)*

### 17. What is the **average `unit_price`**, rounded to 2 decimals?

*מהו המחיר הממוצע ליחידה, מעוגל ל-2 ספרות?*

### 18. How many **null (missing) values** does each column have? Which column is dirty?

*כמה ערכים חסרים (null) יש בכל עמודה? איזו עמודה “מלוכלכת”?*

### 19. Take a **random sample of 5 rows** (use `seed=0` so it is reproducible).

*קחו מדגם אקראי של 5 שורות (השתמשו ב-`seed=0` כדי שיהיה ניתן לשחזור).*

### 20. What is the **estimated in-memory size** of the DataFrame, in MB?

*מהו הגודל המשוער בזיכרון של ה-DataFrame, ב-MB?*

### 21. Show the **5 largest orders by `quantity`** (highest quantity first).

*הציגו את 5 ההזמנות הגדולות ביותר לפי `quantity`.*

### 22. Select rows **10 through 14** (a 5-row slice) using `slice`.
