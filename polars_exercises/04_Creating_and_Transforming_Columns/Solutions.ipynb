# Polars Exercises 04 — Creating & Transforming Columns

This is where transforms come alive. We use `with_columns()` to **add or change** columns: arithmetic (revenue = quantity × price), fixing data types (the string `order_date` → a real date), string cleaning, and conditional logic with `when/then/otherwise`. These are the exact operations we write all day long in Foundry.


**Data file:** `data/orders.csv`

---

### 1. Import Polars as `pl` and read `data/orders.csv` into `orders`.

```python
import polars as pl

orders = pl.read_csv("data/orders.csv")
orders.head()
```

### 2. Add a new column `revenue` = `quantity` × `unit_price`.

```python
orders.with_columns(revenue=pl.col("quantity") * pl.col("unit_price"))
```

### 3. Add `revenue` again, but **rounded to 2 decimals** (use `.round(2)`).

```python
orders.with_columns(
    revenue=(pl.col("quantity") * pl.col("unit_price")).round(2)
)
```

### 4. The `order_date` column is a **string**. Convert it to a real **Date** column (format `"%Y-%m-%d"`).

```python
orders.with_columns(
    pl.col("order_date").str.to_date("%Y-%m-%d")
)
```

### 5. From `order_date`, create a new integer column `order_month` (1–12).

```python
orders.with_columns(
    order_month=pl.col("order_date").str.to_date("%Y-%m-%d").dt.month()
)
```

### 6. Add two columns from `order_date`: `order_year` and `order_quarter`.

```python
d = pl.col("order_date").str.to_date("%Y-%m-%d")
orders.with_columns(
    order_year=d.dt.year(),
    order_quarter=d.dt.quarter(),
)
```

### 7. Add a boolean column `is_return` that is `True` when `quantity` is negative.

```python
orders.with_columns(is_return=pl.col("quantity") < 0)
```

### 8. Add a column `order_size` using `when/then/otherwise`: `"small"` if quantity < 10, `"medium"` if 10–25, `"large"` if > 25.

```python
orders.with_columns(
    order_size=pl.when(pl.col("quantity") < 10).then(pl.lit("small"))
    .when(pl.col("quantity") <= 25).then(pl.lit("medium"))
    .otherwise(pl.lit("large"))
)
```

### 9. Add `region_upper` = the `region` in UPPERCASE (use `str.to_uppercase`).

```python
orders.with_columns(region_upper=pl.col("region").str.to_uppercase())
```

### 10. Add `sku_prefix` = the **first 2 characters** of `product_sku` (use `str.slice`).

```python
orders.with_columns(sku_prefix=pl.col("product_sku").str.slice(0, 2))
```

### 11. Add `channel_clean` where the value `"online"` is replaced by `"web"` (use `str.replace`).

```python
orders.with_columns(
    channel_clean=pl.col("channel").str.replace("online", "web")
)
```

### 12. Add `name_length` = the number of characters in `product_name` (use `str.len_chars`).

```python
orders.with_columns(name_length=pl.col("product_name").str.len_chars())
```

### 13. Fill the missing `discount_code` values with the text `"NONE"` (use `fill_null`).

```python
orders.with_columns(
    discount_code=pl.col("discount_code").fill_null("NONE")
)
```

### 14. Add **two columns at once** in a single `with_columns`: `revenue` (quantity × price) and `is_return` (quantity < 0).

```python
orders.with_columns(
    revenue=pl.col("quantity") * pl.col("unit_price"),
    is_return=pl.col("quantity") < 0,
)
```

### 15. Add `price_with_tax` = `unit_price` × 1.17, rounded to 2 decimals.

```python
orders.with_columns(
    price_with_tax=(pl.col("unit_price") * 1.17).round(2)
)
```

### 16. Build a `segment` column by joining `region` and `channel` with `" / "` between them (use `pl.concat_str`).

```python
orders.with_columns(
    segment=pl.concat_str(["region", "channel"], separator=" / ")
)
```

### 17. Add `priority` = `"high"` when `revenue` (quantity × price) is over 500, otherwise `"normal"`.

```python
orders.with_columns(
    priority=pl.when(pl.col("quantity") * pl.col("unit_price") > 500)
    .then(pl.lit("high"))
    .otherwise(pl.lit("normal"))
)
```

### 18. Add `abs_quantity` = the **absolute value** of `quantity` (returns become positive).

```python
orders.with_columns(abs_quantity=pl.col("quantity").abs())
```
