This short chapter is a fast tour of the basics — but its real goal is to plant the **one idea that makes Spark different**: a DataFrame is not data sitting in your RAM, it is a *recipe* that runs on a cluster, and **nothing runs until you ask for a result** (an *action*).

**Data file:** `data/orders.csv` — 8,000 rows, one row per order line.

### 1. **Setup** — open a local `SparkSession` (already written for you). In Foundry the session is created for you; locally we open one ourselves. `F` is the functions module — almost every column expression lives there.

```python
from pyspark.sql import SparkSession
from pyspark.sql import functions as F

spark = (
    SparkSession.builder
    .appName("spark-course")
    .master("local[*]")
    .config("spark.sql.shuffle.partitions", "8")
    .config("spark.ui.showConsoleProgress", "false")
    .getOrCreate()
)
spark.sparkContext.setLogLevel("ERROR")
print("Spark", spark.version)
```
